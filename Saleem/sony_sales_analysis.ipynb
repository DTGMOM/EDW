# Sony Sales — retail_banking.sony_sales
Loads the full `sony_sales` table from MySQL into a Pandas DataFrame.

```python
import pymysql
import pandas as pd

conn = pymysql.connect(
    host='127.0.0.1',
    port=3306,
    user='root',
    password='Admin123',
    database='retail_banking',
    charset='utf8mb4',
    connect_timeout=10,
)

df = pd.read_sql('SELECT * FROM sony_sales ORDER BY id', conn)
conn.close()

print(f'Total rows loaded: {len(df)}')
df.head(10)
```
